```python


high_school_data = df[df['School Level'] == 'High School']
algebra2_trig_test = 'Algebra2/Trigonometry'
algebra2_trig_data = high_school_data[high_school_data['Regents Exam'] == algebra2_trig_test]

# Check if there is data available for the Algebra2/Trigonometry test
if algebra2_trig_data.empty:
    print(f"No data available for {algebra2_trig_test} test in high schools.")
else:
    # Plot bar chart for Algebra2/Trigonometry scores for each high school
    plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
    plt.bar(algebra2_trig_data['School Name'], algebra2_trig_data['Mean Score'])

    plt.title(f'{algebra2_trig_test} Scores Comparison for High Schools')
    plt.xlabel('High School Name')
    plt.ylabel('Mean Score')
    plt.xticks(rotation=90)  # Rotates the x-axis labels for better readability
    plt.grid(axis='y')
    plt.tight_layout()  # Adjusts the spacing between the plots
    plt.show()




```

```python
print("Unique Test Names:", df2['Regents Exam'].unique())
print("Unique Years:", df2['Year'].unique())

# Filter data to include only high schools and Algebra2/Trigonometry exams for the desired year
desired_year = 2016  # Replace with the year you want to focus on
high_school_data = df2[(df2['School Level'] == 'High School') &
                              (df2['Regents Exam'] == 'Algebra2/Trigonometry') &
                              (df2['Year'] == desired_year)]

# Check if there is data available for the Algebra2/Trigonometry test in the desired year
if high_school_data.empty:
    print(f"No data available for Algebra2/Trigonometry test in high schools for the year {desired_year}.")
else:
    # Plot bar chart for Algebra2/Trigonometry scores for each high school in the desired year
    plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
    plt.bar(high_school_data['School Name'], high_school_data['Mean Score'])

    plt.title(f'Algebra2/Trigonometry Scores Comparison for High Schools ({desired_year})')
    plt.xlabel('High School Name')
    plt.ylabel('Mean Score')
    plt.xticks(rotation=90)  # Rotates the x-axis labels for better readability
    plt.grid(axis='y')
    plt.tight_layout()  # Adjusts the spacing between the plots
    plt.show()
```
